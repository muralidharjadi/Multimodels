```python
import os

OPENAI_API_KEY = "sk"

os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
```

```python
from llama_index.core import SimpleDirectoryReader
from llama_index.multi_modal_llms.openai import OpenAIMultiModal

openai_mm_llm = OpenAIMultiModal(
    model="gpt-4o",
    api_key=OPENAI_API_KEY,
    max_new_tokens=500,
    temperature=0.0,
)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("D:\LlamaIndex\data\image\Verified status.png")
plt.imshow(img)
```

```python
# put your local directory here
image_documents = SimpleDirectoryReader(input_files=["D:\LlamaIndex\data\image\Verified status.png"]).load_data()
```

```python
query = "Analyse the image?"

response_gpt4v = openai_mm_llm.complete(prompt=query,image_documents=image_documents,)

print(response_gpt4v)
```

```python
query = "How many gmail tabs are open?"

response_gpt4v = openai_mm_llm.complete(
    prompt=query,
    image_documents=image_documents,
)

print(response_gpt4v)
```

```python
query = "How many bookmarks are there in this document?"

response_gpt4v = openai_mm_llm.complete(
    prompt=query,
    image_documents=image_documents,
)

print(response_gpt4v)
```
